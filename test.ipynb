```python
# Import necessary libraries
import mss # Screen capturing library
import pytesseract  # Tesseract OCR library for text recognition
from matplotlib import pyplot as plt  # Matplotlib for plotting
from PIL import Image
# Set the Tesseract OCR executable path
pytesseract.pytesseract.tesseract_cmd = 'C:\\Program Files\\Tesseract-OCR\\tesseract.exe'
```

```python
def capture_screen():
    with mss.mss() as sct:
        # Capture the entire screen
        monitor = {"top": 0, "left": 0, "width": 1920, "height": 1080}
        screenshot = sct.grab(monitor)
        return screenshot

def extract_text(image):
    text = pytesseract.image_to_string(image)
    return text

def plot_image(image):
    plt.imshow(image)
    plt.axis('off')
    plt.show()

def main():
    # Capture screen
    screenshot = capture_screen()

    # Convert screenshot to PIL Image
    pil_image = Image.frombytes("RGB", screenshot.size, screenshot.rgb)

    # Extract text from the captured screen
    extracted_text = extract_text(pil_image)
    print("Extracted text:")
    print(extracted_text)

    # Plot the captured screen
    plot_image(pil_image)

if __name__ == "__main__":
    main()
```

```python
# Main function
def main():
    while True:
        # Capture screen
        image = capture_screen()
        
        # Extract text from captured image
        text = extract_text_from_image(image)
        print("Extracted Text:", text)
        
        # Plot the captured screen using Matplotlib
        plt.imshow(image)
        plt.axis('off')  # Turn off axes
        plt.show()

if __name__ == "__main__":
    main()
```
